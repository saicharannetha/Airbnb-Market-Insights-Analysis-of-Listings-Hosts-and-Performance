### **Data Cleaning**

```python
df['name'] = (
    df['name']
    .astype(str)
    .str.replace(r'·.*', '', regex=True)
    .str.strip()
)
```

```python
df['name']
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
numerical_cols = df[['number_of_reviews', 'minimum_nights', 'price', 'calculated_host_listings_count', 'availability_365']]
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
categorical_cols = df[['neighbourhood','room_type','last_review']]
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
for col in numerical_cols:
    df[col] = df[col].astype(int)
```

```python
df.dtypes
```

```python
df['rating'] = df['rating'].replace(['No rating','New '],0)
df['rating'] = df['rating'].astype(float)
```

```python
df['rating']
```

```python
df['bedrooms'] = df['bedrooms'].replace('Studio',0)
df['bedrooms'] = df['bedrooms'].astype(int)
```

```python
df['baths'] = df['baths'].replace('Not specified',0)
df['baths'] = df['baths'].astype(float)
```

### **Handling Outliers**

```python
new_df = df.copy()
```

```python
#calculating Q1, Q3 and IQR
q1 = new_df['price'].quantile(0.25)
q3 = new_df['price'].quantile(0.75)
iqr = q3 - q1

lower_limit = q1 - 1.5 * iqr
upper_limit = q3 + 1.5 * iqr
# capping version of price column
new_df['Price'] = new_df['price'].clip(lower=lower_limit, upper=upper_limit)
```

```python
# finding outliers
outliers = new_df[(new_df['price'] < lower_limit) | (new_df['price'] > upper_limit)]
len(outliers)
```

```python
# skew is balanced 
new_df['Price'].skew()
```

```python
from datetime import datetime

def convert_date(d):
    if pd.isna(d) or d == "":
        return None
    try:
        return datetime.strptime(d, "%d-%m-%Y").strftime("%Y-%m-%d")
    except:
        return None

try:
    insert_query = """
    INSERT INTO airbnb_listings (
        id,
        name,
        host_id,
        host_name,
        neighbourhood_group,
        neighbourhood,
        room_type,
        price,
        minimum_nights,
        number_of_reviews,
        last_review,
        calculated_host_listings_count,
        availability_365,
        license,
        rating,
        bedrooms,
        beds,
        baths,
        license_grouped
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """

    data = [
        (
            float(row['id']),
            row['name'],
            int(row['host_id']),
            row['host_name'],
            row['neighbourhood_group'],
            row['neighbourhood'],
            row['room_type'],
            int(row['price']),
            int(row['minimum_nights']),
            int(row['number_of_reviews']),
            convert_date(row['last_review']),
            int(row['calculated_host_listings_count']),
            int(row['availability_365']),
            row['license'],
            float(row['rating']),
            int(row['bedrooms']),
            int(row['beds']),
            float(row['baths']),
            row['license_grouped']
        )
        for index, row in df.iterrows()
    ]

    cursor.executemany(insert_query, data)
    conn.commit()
    print("Data inserted successfully")

except mysql.connector.Error as err:
    print("Error:", err)
```
